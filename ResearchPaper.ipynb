```python
cost_table = cost_table.apply(pd.to_numeric, errors='coerce')
exec_table = exec_table.apply(pd.to_numeric, errors='coerce')
```

```python
# Paramètres du problème
n_tasks = task_details.shape[0]
n_nodes = node_details.shape[0]

# Pour combiner les deux objectifs, nous utilisons une somme pondérée.
# Vous pouvez ajuster ces coefficients en fonction de l'importance relative de chaque objectif.
alpha = 1  # pondération du coût
beta = 0  # pondération du temps d'exécution
```

```python
print("CostTable shape:", cost_table.shape)
print("ExecutionTable shape:", exec_table.shape)
print("Nombre de nœuds:", n_nodes, "Nombre de tâches:", n_tasks)
```

```python
def evaluate_solution(assignment):
    """
    Calcule l'objectif combiné pour une solution donnée (vecteur d'assignation de taille n_tasks).
    assignment : tableau d'entiers, chaque valeur est dans [0, n_nodes-1].

    On suppose ici que dans CostTable et ExecutionTable, les lignes représentent les nœuds
    et les colonnes représentent les tâches.
    """
    total_cost = 0.0
    node_exec_times = np.zeros(n_nodes)

    for t in range(n_tasks):
        node = assignment[t]
        total_cost += cost_table.iloc[node, t]
        node_exec_times[node] += exec_table.iloc[node, t]

    total_exec_time = np.max(node_exec_times)  # le temps total est dominé par le nœud le plus chargé
    return alpha * total_cost + beta * total_exec_time

def HHO(max_iter=100, pop_size=30):
    # Initialisation de la population : chaque solution est un vecteur de taille n_tasks
    # avec des indices aléatoires entre 0 et n_nodes-1
    population = [np.random.randint(0, n_nodes, n_tasks) for _ in range(pop_size)]
    fitness = np.array([evaluate_solution(sol) for sol in population])

    best_idx = np.argmin(fitness)
    best_solution = population[best_idx].copy()
    best_fitness = fitness[best_idx]

    E0 = 2  # énergie initiale
    for t in range(max_iter):
        E = E0 * (1 - t / max_iter)  # énergie décroissante

        for i in range(pop_size):
            current_sol = population[i].copy()

            # Phase d'exploration/exploitation en fonction de l'énergie E
            if abs(E) >= 1:
                # Exploration : se rapprocher d'une solution aléatoire
                q = np.random.rand()
                rand_sol = population[np.random.randint(0, pop_size)]
                new_sol = current_sol + q * (rand_sol - current_sol)
            else:
                # Exploitation
                r = np.random.rand()
                if r >= 0.5:
                    new_sol = best_solution - E * abs(best_solution - current_sol)
                else:
                    new_sol = best_solution + E * (np.random.rand(n_tasks) - 0.5)

            # Discrétisation de la solution candidate
            new_sol = np.rint(new_sol).astype(int)
            new_sol = np.clip(new_sol, 0, n_nodes - 1)

            new_fitness = evaluate_solution(new_sol)
            if new_fitness < fitness[i]:
                population[i] = new_sol.copy()
                fitness[i] = new_fitness
                if new_fitness < best_fitness:
                    best_solution = new_sol.copy()
                    best_fitness = new_fitness

        print(f"Iter {t+1}/{max_iter} - Meilleur fitness: {best_fitness:.4f}")

    return best_solution, best_fitness

# Exécution de l'algorithme HHO
best_sol, best_fit = HHO(max_iter=200, pop_size=30)

print("\nMeilleure assignation trouvée :")
print(best_sol)
print("Fitness associée :", best_fit)
```

```python
def evaluate_solution(assignment):
    """
    Calcule l'objectif combiné pour une solution donnée (vecteur d'assignation de taille n_tasks).
    assignment : tableau d'entiers, chaque valeur est dans [0, n_nodes-1].

    On suppose ici que dans CostTable et ExecutionTable, les lignes représentent les nœuds
    et les colonnes représentent les tâches.
    """
    total_cost = 0.0
    node_exec_times = np.zeros(n_nodes)

    for t in range(n_tasks):
        node = assignment[t]
        total_cost += cost_table.iloc[node, t]
        node_exec_times[node] += exec_table.iloc[node, t]

    total_exec_time = np.max(node_exec_times)  # le temps total est dominé par le nœud le plus chargé
    return alpha * total_cost + beta * total_exec_time

import math

def levy_flight(dimension):
    """
    Génère un mouvement selon une distribution de Lévy pour améliorer l'exploration.
    """
    beta = 1.5
    sigma = (math.gamma(1 + beta) * math.sin(np.pi * beta / 2) /
             (math.gamma((1 + beta) / 2) * beta * 2 ** ((beta - 1) / 2))) ** (1 / beta)
    u = np.random.normal(0, sigma, dimension)
    v = np.random.normal(0, 1, dimension)
    step = u / np.power(np.abs(v), 1 / beta)
    return 0.01 * step

def improved_HHO(max_iter=100, pop_size=30, no_improve_limit=30, local_search=True):
    """
    Version améliorée de l'algorithme Harris Hawks Optimization.

    Paramètres:
    - cost_table, exec_table: tables des coûts et temps d'exécution
    - n_tasks, n_nodes: nombre de tâches et de nœuds
    - alpha, beta: paramètres de pondération pour le coût et le temps d'exécution
    - max_iter: nombre maximum d'itérations
    - pop_size: taille de la population
    - no_improve_limit: nombre d'itérations sans amélioration avant redémarrage
    - local_search: activer la recherche locale
    """
    # Initialisation de la population avec une stratégie diversifiée
    population = []
    for i in range(pop_size):
        if i < pop_size // 3:
            # Assignation aléatoire complète
            sol = np.random.randint(0, n_nodes, n_tasks)
        elif i < 2 * (pop_size // 3):
            # Assignation par regroupement (clustering)
            sol = np.zeros(n_tasks, dtype=int)
            for t in range(n_tasks):
                # Assigne chaque tâche au nœud avec le meilleur rapport coût/temps
                ratios = np.array([cost_table.iloc[n, t] / max(0.1, exec_table.iloc[n, t])
                                  for n in range(n_nodes)])
                sol[t] = np.argmin(ratios)
        else:
            # Assignation équilibrée par charge
            sol = np.zeros(n_tasks, dtype=int)
            node_loads = np.zeros(n_nodes)
            for t in range(n_tasks):
                # Trouve le nœud le moins chargé
                node = np.argmin(node_loads)
                sol[t] = node
                node_loads[node] += exec_table.iloc[node, t]

        population.append(sol)

    fitness = np.array([evaluate_solution(sol)
                        for sol in population])

    best_idx = np.argmin(fitness)
    best_solution = population[best_idx].copy()
    best_fitness = fitness[best_idx]

    # Historique de convergence
    history = [best_fitness]
    best_iter = 0

    E0 = 2.0  # énergie initiale

    for t in range(max_iter):
        E = E0 * (1 - t / max_iter)  # énergie décroissante

        # Détection de stagnation et redémarrage partiel
        if t - best_iter > no_improve_limit:
            print(f"Redémarrage à l'itération {t+1} - Stagnation détectée")
            # Garder les meilleures solutions (10%)
            elite_count = max(1, pop_size // 10)
            elite_indices = np.argsort(fitness)[:elite_count]

            # Régénérer 90% de la population
            for i in range(pop_size):
                if i not in elite_indices:
                    if np.random.rand() < 0.5:
                        # Nouvelle solution aléatoire
                        population[i] = np.random.randint(0, n_nodes, n_tasks)
                    else:
                        # Perturbation de la meilleure solution
                        perturb = np.random.randint(0, n_nodes, n_tasks)
                        mask = np.random.rand(n_tasks) < 0.3  # 30% des positions
                        new_sol = best_solution.copy()
                        new_sol[mask] = perturb[mask]
                        population[i] = new_sol

                    fitness[i] = evaluate_solution(population[i])

        for i in range(pop_size):
            current_sol = population[i].copy()

            # Sélection aléatoire entre plusieurs stratégies
            strategy = np.random.randint(0, 4)

            if strategy == 0:
                # Exploration avec vol de Lévy
                levy_step = levy_flight(n_tasks)
                new_sol = current_sol.copy().astype(float)
                for j in range(n_tasks):
                    if np.random.rand() < 0.3:  # Modification partielle
                        new_sol[j] += levy_step[j]

            elif strategy == 1 and abs(E) >= 1:
                # Exploration standard: se rapprocher d'une solution aléatoire
                q = np.random.rand()
                rand_idx = np.random.randint(0, pop_size)
                rand_sol = population[rand_idx]
                new_sol = current_sol + np.random.rand(n_tasks) * (rand_sol - current_sol)

            elif strategy == 2:
                # Exploitation basée sur la meilleure solution
                r = np.random.rand()
                if r >= 0.5:
                    new_sol = best_solution - E * abs(best_solution - current_sol)
                else:
                    # Recherche plus agressive autour de la meilleure solution
                    new_sol = best_solution + E * np.random.randn(n_tasks)

            else:
                # Stratégie d'exploitation améliorée - mouvements ciblés
                new_sol = current_sol.copy()

                # Identifier les tâches les plus coûteuses
                task_costs = np.array([cost_table.iloc[current_sol[t], t] for t in range(n_tasks)])
                worst_tasks = np.argsort(-task_costs)[:max(1, int(0.2 * n_tasks))]  # 20% pires tâches

                for task in worst_tasks:
                    # Tester plusieurs nœuds alternatifs pour cette tâche
                    current_node = current_sol[task]
                    best_node = current_node
                    best_task_fitness = float('inf')

                    # Vérifier seulement quelques nœuds aléatoires pour efficacité
                    candidate_nodes = np.random.choice(n_nodes, min(5, n_nodes), replace=False)

                    for node in candidate_nodes:
                        if node != current_node:
                            test_sol = new_sol.copy()
                            test_sol[task] = node
                            test_fitness = evaluate_solution(test_sol)
                            if test_fitness < best_task_fitness:
                                best_task_fitness = test_fitness
                                best_node = node

                    new_sol[task] = best_node

            # Discrétisation de la solution candidate
            new_sol = np.rint(new_sol).astype(int)
            new_sol = np.clip(new_sol, 0, n_nodes - 1)

            # Recherche locale pour améliorer la solution (si activée)
            if local_search and np.random.rand() < 0.1:  # 10% de chance
                for _ in range(min(5, n_tasks)):  # Limiter le nombre d'essais
                    task = np.random.randint(0, n_tasks)
                    current_node = new_sol[task]
                    for node in range(n_nodes):
                        if node != current_node:
                            test_sol = new_sol.copy()
                            test_sol[task] = node
                            test_fitness = evaluate_solution(test_sol)
                            if test_fitness < evaluate_solution(new_sol):
                                new_sol = test_sol.copy()

            new_fitness = evaluate_solution(new_sol)

            # Mise à jour avec acceptation probabiliste pour éviter les optima locaux
            if new_fitness < fitness[i]:
                # Acceptation directe si amélioration
                population[i] = new_sol.copy()
                fitness[i] = new_fitness

                # Mise à jour de la meilleure solution globale
                if new_fitness < best_fitness:
                    best_solution = new_sol.copy()
                    best_fitness = new_fitness
                    best_iter = t
            elif np.random.rand() < 0.05 * np.exp(-(new_fitness - fitness[i]) / (max(1, t) * 0.01)):
                # Acceptation avec faible probabilité même si dégradation (simulated annealing)
                population[i] = new_sol.copy()
                fitness[i] = new_fitness

        # Enregistrement de l'histoire
        history.append(best_fitness)

        if (t + 1) % 10 == 0 or t == 0 or t == max_iter - 1:
            print(f"Iter {t+1}/{max_iter} - Meilleur fitness: {best_fitness:.4f}")

    # Analyse de la charge des nœuds pour la meilleure solution
    node_loads = np.zeros(n_nodes)
    for t in range(n_tasks):
        node = best_solution[t]
        node_loads[node] += exec_table.iloc[node, t]

    print("\nDistribution des charges des nœuds:")
    for n in range(n_nodes):
        print(f"Nœud {n}: {node_loads[n]:.2f} unités de temps, {np.sum(best_solution == n)} tâches")

    # Calcul des métriques spécifiques
    total_cost = sum(cost_table.iloc[best_solution[t], t] for t in range(n_tasks))
    max_exec_time = np.max(node_loads)

    print(f"\nCoût total: {total_cost:.2f}")
    print(f"Temps d'exécution maximal: {max_exec_time:.2f}")
    print(f"Fonction objectif: {best_fitness:.4f} (alpha={alpha}, beta={beta})")


    return best_solution, best_fitness, history

# Exemple d'utilisation:
best_sol, best_fit, convergence = improved_HHO(max_iter=100, pop_size=50,
                                              no_improve_limit=30, local_search=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import levy
from mpl_toolkits.mplot3d import Axes3D
import math


def evaluate_solution_multi_objective(assignment):
    """
    Calcule les deux objectifs séparément pour une solution donnée.
    Retourne un tuple (coût total, temps d'exécution maximal).
    """
    total_cost = 0.0
    node_exec_times = np.zeros(n_nodes)

    for t in range(n_tasks):
        node = assignment[t]
        total_cost += cost_table.iloc[node, t]
        node_exec_times[node] += exec_table.iloc[node, t]

    total_exec_time = np.max(node_exec_times)

    return total_cost, total_exec_time

def levy_flight(dim):
    """
    Génère un pas de vol de Lévy pour améliorer l'exploration.
    """
    beta = 1.5  # Paramètre de Lévy (1 < beta < 2)
    sigma = (math.gamma(1 + beta) * math.sin(np.pi * beta / 2) /
             (math.gamma((1 + beta) / 2) * beta * 2**((beta - 1) / 2)))**(1 / beta)

    u = np.random.normal(0, sigma, dim)
    v = np.random.normal(0, 1, dim)
    step = u / abs(v)**(1 / beta)

    return step

def is_dominated(p1, p2):
    """
    Vérifie si p1 est dominé par p2 (pour minimisation).
    """
    return (p1[0] >= p2[0] and p1[1] >= p2[1]) and (p1[0] > p2[0] or p1[1] > p2[1])

def get_pareto_front(solutions, fitnesses):
    """
    Identifie le front de Pareto parmi toutes les solutions.
    """
    pareto_indices = []

    for i in range(len(fitnesses)):
        dominated = False
        for j in range(len(fitnesses)):
            if i != j and is_dominated(fitnesses[i], fitnesses[j]):
                dominated = True
                break

        if not dominated:
            pareto_indices.append(i)

    return [solutions[i] for i in pareto_indices], [fitnesses[i] for i in pareto_indices]

def HHO_multi_objective(max_iter=100, pop_size=30):
    # Initialisation de la population
    population = [np.random.randint(0, n_nodes, n_tasks) for _ in range(pop_size)]

    # Évaluation multi-objectif
    fitness_values = [evaluate_solution_multi_objective(sol) for sol in population]

    # Stockage des solutions pour visualisation
    all_solutions = []
    for sol, fit in zip(population, fitness_values):
        all_solutions.append((sol.copy(), fit))

    # Initialisation du front de Pareto
    pareto_solutions, pareto_fitness = get_pareto_front(population, fitness_values)

    # Sélection d'une solution "leader" pour guider la recherche
    # On choisit arbitrairement la première solution du front de Pareto
    if pareto_solutions:
        best_solution = pareto_solutions[0].copy()
    else:
        # Utiliser une solution aléatoire si le front est vide
        best_solution = population[0].copy()

    E0 = 2  # énergie initiale
    for t in range(max_iter):
        E = E0 * (1 - t / max_iter)  # énergie décroissante

        for i in range(pop_size):
            current_sol = population[i].copy()

            # Phase d'exploration/exploitation en fonction de l'énergie E
            if abs(E) >= 1:
                # Exploration : se rapprocher d'une solution aléatoire
                q = np.random.rand()
                rand_sol = population[np.random.randint(0, pop_size)]
                new_sol = current_sol + np.random.rand(n_tasks) * (rand_sol - current_sol)
            else:
                # Exploitation
                r = np.random.rand()
                if r >= 0.5:
                    new_sol = best_solution - E * abs(best_solution - current_sol)
                else:
                    # Ajout de Lévy flight pour éviter les optima locaux
                    levy_step = levy_flight(n_tasks)
                    new_sol = best_solution + E * levy_step

            # Discrétisation de la solution candidate
            new_sol = np.rint(new_sol).astype(int)
            new_sol = np.clip(new_sol, 0, n_nodes - 1)

            # Évaluation multi-objectif
            new_fitness = evaluate_solution_multi_objective(new_sol)

            # Mise à jour basée sur la dominance de Pareto
            current_fitness = fitness_values[i]

            # Si la nouvelle solution domine l'ancienne ou n'est pas dominée par l'ancienne
            if is_dominated(current_fitness, new_fitness) or not is_dominated(new_fitness, current_fitness):
                population[i] = new_sol.copy()
                fitness_values[i] = new_fitness

                # Stocker la nouvelle solution
                all_solutions.append((new_sol.copy(), new_fitness))

        # Mise à jour du front de Pareto
        pareto_solutions, pareto_fitness = get_pareto_front(population, fitness_values)

        # Choisir aléatoirement une solution du front de Pareto comme leader
        if pareto_solutions:
            best_solution = pareto_solutions[np.random.randint(0, len(pareto_solutions))].copy()

        print(f"Iter {t+1}/{max_iter} - Taille du front de Pareto: {len(pareto_fitness)}")

    return pareto_solutions, pareto_fitness, all_solutions

def plot_solutions(all_solutions, pareto_front_data):
    """
    Visualise toutes les solutions et le front de Pareto.

    all_solutions: liste de tuples (solution, (coût, temps))
    pareto_front_data: liste de tuples (coût, temps)
    """
    # Extraire les coûts et temps d'exécution de toutes les solutions
    costs = [sol[1][0] for sol in all_solutions]
    times = [sol[1][1] for sol in all_solutions]

    plt.figure(figsize=(10, 8))
    plt.scatter(costs, times, alpha=0.5, label='Solutions explorées')

    # Traitement du front de Pareto
    if pareto_front_data:
        pareto_costs = [p[0] for p in pareto_front_data]
        pareto_times = [p[1] for p in pareto_front_data]

        plt.scatter(pareto_costs, pareto_times, color='red', marker='*', s=100, label='Front de Pareto')

        # Connecter les points du front de Pareto
        pareto_points = sorted(zip(pareto_costs, pareto_times), key=lambda x: x[0])
        pareto_costs_sorted = [p[0] for p in pareto_points]
        pareto_times_sorted = [p[1] for p in pareto_points]
        plt.plot(pareto_costs_sorted, pareto_times_sorted, 'r--')

    plt.xlabel('Coût total')
    plt.ylabel('Temps d\'exécution maximal')
    plt.title('Espace des objectifs et front de Pareto')
    plt.legend()
    plt.grid(True)
    plt.show()

# Exécution de l'algorithme HHO multi-objectif
pareto_solutions, pareto_fitness, all_solutions = HHO_multi_objective(max_iter=50, pop_size=30)

# Affichage des résultats
print(f"\nNombre de solutions dans le front de Pareto: {len(pareto_solutions)}")
print("\nSolutions du front de Pareto (coût, temps):")
for i, (sol, fit) in enumerate(zip(pareto_solutions, pareto_fitness)):
    print(f"Solution {i+1}: Coût = {fit[0]:.2f}, Temps = {fit[1]:.2f}")

# Si vous souhaitez obtenir une solution spécifique du front de Pareto
# Par exemple, celle avec le meilleur compromis, on peut utiliser la méthode du point idéal
if pareto_fitness:
    # Normalisation des objectifs
    min_cost = min(f[0] for f in pareto_fitness)
    max_cost = max(f[0] for f in pareto_fitness)
    min_time = min(f[1] for f in pareto_fitness)
    max_time = max(f[1] for f in pareto_fitness)

    # Point idéal (normalisé)
    ideal_point = [0, 0]

    # Trouver la solution la plus proche du point idéal
    best_idx = 0
    best_distance = float('inf')

    for i, fit in enumerate(pareto_fitness):
        # Normalisation
        norm_cost = (fit[0] - min_cost) / (max_cost - min_cost) if max_cost > min_cost else 0
        norm_time = (fit[1] - min_time) / (max_time - min_time) if max_time > min_time else 0

        # Distance euclidienne au point idéal
        distance = np.sqrt((norm_cost - ideal_point[0])**2 + (norm_time - ideal_point[1])**2)

        if distance < best_distance:
            best_distance = distance
            best_idx = i

    print("\nMeilleure solution de compromis:")
    print("Assignation:", pareto_solutions[best_idx])
    print(f"Coût: {pareto_fitness[best_idx][0]:.2f}, Temps: {pareto_fitness[best_idx][1]:.2f}")
```
